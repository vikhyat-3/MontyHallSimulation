```python
import math
import random
import matplotlib.pyplot as plt
def createMontyHallSimulation(n):
  switchSuccess,dontSwitchSuccess=[],[]
  #where n is the number of simulation per number of doors
  for k in range(3,50): #k is the number of closed doors present
    switch,dontSwitch=0,0
    for i in range(n): #looping through each simulation scenario

      #randomly choosing the door the contestant chooses
      chosenDoor=random.randint(1,k) 
      
      #randomly choosing the door behind which the prize actually exists
      prizeDoor=random.randint(1,k)

      #initalizing the door to be revealed by Monty
      revealDoor=random.randint(1,k)
      #Ensuring that the revealed door is not the one chosen by the contestant, nor is it the one which contains the prize
      while revealDoor==prizeDoor or revealDoor==chosenDoor:
        revealDoor=random.randint(1,k)
        
      #initalizing the door to be switched to by the contestant
      changedDoor=random.randint(1,k)
      #Ensuring that the switched door is not the one chosen initally by the contestant, nor is it the one which has been revealed by Monty
      while changedDoor==chosenDoor or changedDoor==revealDoor:
        changedDoor=random.randint(1,k)

      #If the switched door contains the prize, increment counter 'switch'
      switch+=1 if changedDoor==prizeDoor else 0
      #If the initally chosen door by the contestant contains the prize, increment counter 'dontSwitch'
      dontSwitch+=1 if chosenDoor==prizeDoor else 0
    switchSuccess.append(switch*100/n)
    dontSwitchSuccess.append(dontSwitch*100/n)
  return switchSuccess,dontSwitchSuccess
```

```python
switchSuccessRate,dontSwitchSucessRate=createMontyHallSimulation(5000)
for doorCount,(i,j) in enumerate(zip(switchSuccessRate,dontSwitchSucessRate)):
  print(doorCount+3,"doors: ",i,"%",j,"%") #Printing the chances of winning when the contestant decides to switch and not to switch
```

```python
x_values = range(1, len(switchSuccessRate) + 1)

# Plotting the lines
plt.plot(x_values, switchSuccessRate, label='Array 1', marker='o')
plt.plot(x_values, dontSwitchSucessRate, label='Array 2', marker='o')

# Adding labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Graph of Arrays')
plt.yticks(range(0, math.ceil(max(max(switchSuccessRate), max(dontSwitchSucessRate))) + 1, 5))
plt.legend()  # Show legend

# Display the plot
plt.show()
```
